# Preprocessing data and model training

Import dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load Dataset

```python
df=pd.read_csv('matplotlib_data.csv')
df.head()
```

Create a line chart to analyze the sales

```python
df['Sales'].plot(kind='line',title='Line plot of Sales',color='grey') 
plt.xlabel('Index')
plt.ylabel('Sales')
plt.show()
```

Create a plot ot compare average sales per category.

```python
#Bar plot example
df.groupby('Category')['Sales'].mean().plot(kind='bar',title='Average Sales per Category')
plt.xlabel('Category')
plt.ylabel('Average Sales')
plt.legend()
plt.show()
```

Create a plot to compare the distribution of data per category

```python
#Pie chart example
df['Category'].value_counts().plot(kind='pie',autopct='%1.1f%%',title='Category Distribution')
plt.ylabel('') #remove extra label from y-axis
plt.legend()
plt.show()
```

Create a histogram for profit column

```python
#Histogram example
df['Profit'].plot(kind='hist',bins=20,title='Histogram of Profit',alpha=0.7)
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Create two sub-plots side by side to find the patterns for sales and profit

```python
#fig -> overall figure object (canvas)
#axs -> array of axes (sub-plots)
fig,axs=plt.subplots(1,2,figsize=(10,4)) #1 row, 2 columns
axs[0].plot(df['Sales'],color='blue',linestyle=':')
axs[0].set_title('Sales')
axs[1].plot(df['Profit'],color='black',linestyle=':')
axs[1].set_title('Profit')
plt.tight_layout()
plt.show()
```

Create four sub-plots like 2x2 representation

```python
fig,axs=plt.subplots(2,2,figsize=(8,8)) #1 row, 2 columns
axs[0,0].plot(df['Sales'],linestyle='-')
axs[0,0].set_title('Sales')
axs[0,1].plot(df['Profit'],color='black',linestyle='-')
axs[0,1].set_title('Profit')
axs[1,0].plot(df['Sales'],color='green',linestyle='-')
axs[1,0].set_title('Sales')
axs[1,1].plot(df['Profit'],color='red',linestyle='-')
axs[1,1].set_title('Profit')
plt.tight_layout() 
plt.show() 
```
