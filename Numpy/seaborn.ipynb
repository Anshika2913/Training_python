```python
##import libraries and load dataset
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load sample dataset from seaborn
df = sns.load_dataset('titanic')
```

```python
#display first 5 rows of the dataset
print("Sample Data:")
df.head()
```

```python
#data information
print("\nData Information:")
print(df.info())
```

```python
#summary statistics of numerical columns
print("\nSummary statistics of numerical columns:")
df.describe()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='age', y='fare')
plt.title('Scatter Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
#scatter plot - Visualizing the relationship between age and fare
#scatter plots are used to visualize the relationship between two numerical variables.
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='age', y='fare', hue='survived',size='adult_male')
plt.title('Scatter Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
#limitation of hue variable in scatter plot
# The hue parameter can only represent one categorical variable at a time, limiting the ability to visualize multiple categories simultaneously.
```

```python
#count plot - Visualizing the distribution of passengers by class
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='class', hue='who')
plt.title('Count Plot: Passenger Class Distribution')
plt.legend(title='Who', loc='upper right')
plt.show()
```

```python
#continous vs categorical data
#box plots are used to visualize the distribution of a numerical variable across different categories of a categorical variable.
#continouous data- mean or median is used to represent the central tendency of the data.
#categorical data- mode is used to represent the most frequent category in the data.
```

```python
#Box plot
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='class', y='age', hue='sex')
plt.title('Box Plot: Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend(title='sex', loc='upper right')
plt.show()
```

```python
#Swarm plot
plt.figure(figsize=(7,5))
sns.swarmplot(data=df, x='class', y='age', hue='sex')
plt.title('Swarm Plot: Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
#heatmap
#heatmaps are used to visualize the correlation between numerical variables in a dataset.
plt.figure(figsize=(4,4))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
# sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap: Correlation Matrix')
plt.show()
```

```python
#Pair plot
sns.pairplot(df,hue='sex')
plt.suptitle('Pair Plot: Numerical Features', y=1.02)
plt.show()
```
